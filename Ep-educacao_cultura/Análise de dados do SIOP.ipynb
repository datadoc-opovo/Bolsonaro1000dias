```python
presidentes = ['Lula','Lula2','Dilma','Bolsonaro']
ano_posse = [2003,2007,2011,2019]


df_presidentes = pd.DataFrame(ano_posse, index = presidentes, columns = ['Ano Posse'])

df_presidentes['Ano Anterior'] = df_presidentes['Ano Posse']-1

df_presidentes['Segundo Ano'] = df_presidentes['Ano Posse']+1

df_presidentes = df_presidentes.reset_index()

df_presidentes
```

```python
grupo_ano.loc[2002]
```

```python
grupo_ano.loc[2004]
```

```python
merge_1 = pd.merge(df_presidentes,grupo_ano,left_on='Ano Anterior',right_on='Ano')
merge_1
```

```python
resumo_orc_presidente = pd.merge(merge_1,grupo_ano,left_on='Segundo Ano',right_on='Ano')
resumo_orc_presidente
```

```python
resumo_orc_presidente.columns
```

```python
resumo_orc_presidente.columns = ['Presidente','Ano Posse', 'Ano Anterior', 'Segundo Ano', 'orçamento_ano_anterior',
       'executado_ano_anterior', 'orçamento_segundo_ano', 'executado_segundo_ano']
```

```python
resumo_orc_presidente
```

```python
resumo_orc_presidente['Variação (%) Orçamento'] = (resumo_orc_presidente['orçamento_segundo_ano']/resumo_orc_presidente['orçamento_ano_anterior']-1) * 100
resumo_orc_presidente['Variação (%) Executado'] = (resumo_orc_presidente['executado_segundo_ano']/resumo_orc_presidente['executado_ano_anterior']-1) * 100
```

```python
df_2018 = selecao.loc[(selecao.Ano == 2018)]
df_2020 = selecao.loc[(selecao.Ano == 2020)]
```

```python
gp18MA = df_2018.groupby(['Modalidade de Aplicação']).sum().reset_index().sort_values(by=['Executado'], ascending=False).filter(['Modalidade de Aplicação','Executado'])
gp20MA = df_2020.groupby(['Modalidade de Aplicação']).sum().reset_index().sort_values(by=['Executado'], ascending=False).filter(['Modalidade de Aplicação','Executado'])
```

```python
df_MA = pd.merge(gp18,gp20,how='outer',on='Modalidade de Aplicação')
df_MA
```

```python
gp18Sb = df_2018.groupby(['Subfunção']).sum().reset_index().sort_values(by=['Executado'], ascending=False).filter(['Subfunção','Executado'])
gp20Sb = df_2020.groupby(['Subfunção']).sum().reset_index().sort_values(by=['Executado'], ascending=False).filter(['Subfunção','Executado'])
```

```python
df_Sb = pd.merge(gp18Sb,gp20Sb,how='outer',on='Subfunção')
df_Sb
```

```python
gp18UO = df_2018.groupby(['Unidade Orçamentária']).sum().reset_index().sort_values(by=['Executado'], ascending=False).filter(['Unidade Orçamentária','Executado'])
gp20UO = df_2020.groupby(['Unidade Orçamentária']).sum().reset_index().sort_values(by=['Executado'], ascending=False).filter(['Unidade Orçamentária','Executado'])
```

```python
df_UO = pd.merge(gp18UO,gp20UO,how='outer',on='Unidade Orçamentária')
df_UO
```
